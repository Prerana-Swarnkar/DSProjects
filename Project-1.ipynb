```python
import numpy as np
import pandas as pd
#read csv file
ds_titanic=pd.read_csv("titanic.csv")
ds_titanic
```

```python
#put data set into data frame
df_titanic=pd.DataFrame(data=ds_titanic)
df_titanic
```

```python
#check the null values using heat map
import seaborn as sb
import matplotlib.pyplot as plt
sb.heatmap(df_titanic.isnull(),annot=True)
plt.show()
```

```python
# import imput class from sklearn library for imputing null values 
from sklearn.impute import SimpleImputer

# create instance of imputer class
imp = SimpleImputer(strategy='most_frequent')
df_titanic['Age']=imp.fit_transform(df_titanic['Age'].values.reshape(-1,1))
df_titanic['Cabin']=imp.fit_transform(df_titanic['Cabin'].values.reshape(-1,1))
df_titanic
```

```python
#use LabelEncoding Tachnic to change Sex column values and embarked
from sklearn.preprocessing import LabelEncoder
le_titanic= LabelEncoder()
columns=['Sex','Embarked','Cabin','Ticket']
for val in columns:
    df_titanic[val]=le_titanic.fit_transform(df_titanic[val].astype(str))
df_titanic
```

```python
# identify skewness
df_titanic.skew()
```

```python
x = df_titanic.drop('Name',axis=1)
x
```

```python
df_titanic_new=pd.DataFrame(data=x)
df_titanic_new
```

```python
#skewness treating
for columns in df_titanic_new.columns:
    if df_titanic_new.skew().loc[columns]>0.55:
        df_titanic_new[columns]=np.log1p(df_titanic[columns])
df_titanic_new
```

```python
#spliting data
X=df_titanic_new.drop('Survived',axis=1)
X
```

```python
Y=df_titanic_new['Survived']
Y
```

```python
X.shape
```

```python
Y.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=.30,random_state=50)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
Y_test.shape
```

```python
Y_test.shape
```

```python
lg=LogisticRegression()
```

```python
lg.fit(X_train,Y_train)
```

```python
lg.score(X_train,Y_train)
```

```python
pred=lg.predict(X_test)
print(pred)
```

```python
print('accuracy_score:',accuracy_score(Y_test,pred))
```

```python
print (confusion_matrix(Y_test,pred))
```

```python
print(classification_report(Y_test,pred))
```
