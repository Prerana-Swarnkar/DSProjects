```python
from sklearn.preprocessing import LabelEncoder
le_sonar= LabelEncoder()
sonar_data['R']=le_sonar.fit_transform(sonar_data['R'].astype(str))
sonar_data
```

```python
for i in sonar_data.columns:
    plt.figure()
    sns.distplot(sonar_data[i])
```

```python
from scipy.stats import zscore
z_score=abs(zscore(sonar_data))
print(sonar_data.shape)
sonar_data_final=sonar_data.loc[(z_score<3)]
print(sonar_data_final.shape)
```

```python
sonar_data.skew()
```

```python
dst=pd.DataFrame(sonar_data_final)
dst
```

```python
y=dst.iloc[:,-1]
y.head()
```

```python
y.shape
```

```python
x=dst.iloc[:,0:-1]
```

```python
x.head()
```

```python
x.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=100)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
lg=LogisticRegression()
```

```python
lg.fit(x_train,y_train)
```

```python
lg.score(x_train,y_train)
```

```python
pred=lg.predict(x_test)
print(pred)
```

```python
print('accuracy_score:',accuracy_score(y_test,pred))
```

```python
print (confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```
